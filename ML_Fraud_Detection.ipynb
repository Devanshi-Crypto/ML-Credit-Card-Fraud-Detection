```python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo
import numpy as np
import joblib


root = tk.Tk()
root.title("Credit Card Fraud Detection System")
root.geometry("600x600")

# Set background color
root.configure(bg='blue')

# Create a canvas widget for the main frame with scrolling
canvas = tk.Canvas(root)
canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
scrollbar = ttk.Scrollbar(root, orient=tk.VERTICAL, command=canvas.yview)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
canvas.configure(yscrollcommand=scrollbar.set)

# Create a frame inside the canvas for the main frame
main_frame = ttk.Frame(canvas)
canvas.create_window((0, 0), window=main_frame, anchor='nw')

# Set background color and highlight color
style = ttk.Style()
style.configure('My.TFrame', background='light blue', borderwidth=5)
main_frame.configure(style='My.TFrame')

# Header
header_frame = ttk.Frame(main_frame)
header_frame.pack(pady=20)
header = ttk.Label(header_frame, text="Credit Card Fraud Detection System", font=("Arial", 20))
header.pack()

# Inputs
inputs_frame = ttk.Frame(main_frame)
inputs_frame.pack(pady=20)

inputs = [
    "V1", "V2", "V3", "V4", "V5", "V6",
    "V7", "V8", "V9", "V10", "V11", "V12",
    "V13", "V14", "V15", "V16", "V17", "V18",
    "V19", "V20", "V21", "V22", "V23", "V24",
    "V25", "V26", "V27", "V28", "Amount"
]

entries = {}
for i, inp in enumerate(inputs):
    ttk.Label(inputs_frame, text=f"Enter value of {inp}").grid(column=0, row=i)
    entries[inp] = ttk.Entry(inputs_frame)
    entries[inp].grid(column=1, row=i, padx=10)

def detect_fraud():
    # Load the machine learning model
    joblib.dump(rf1,"credit_card_model")
    model = joblib.load("credit_card_model")
    
    # Convert user inputs to a list of floats
    input_values = []
    for inp in inputs:
        try:
            input_values.append(float(entries[inp].get()))
        except ValueError:
            showinfo("Error", f"Invalid value for {inp}")
            return
    
    # Make prediction
    y_pred = model.predict([input_values])
    
    # Display prediction result
    if y_pred == 0:
        showinfo("Result", "Normal Transaction")
    else:
        showinfo("Result", "Fraudulent Transaction")

# Predict button
button_frame = ttk.Frame(main_frame)
button_frame.pack(pady=20)
predict_button = ttk.Button(button_frame, text="Predict", command=detect_fraud)
predict_button.pack()

# Update the canvas scroll region
main_frame.update_idletasks()
canvas.configure(scrollregion=canvas.bbox('all'))

root.mainloop()
```
